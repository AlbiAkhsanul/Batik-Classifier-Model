```python
import os

# Ganti ini dengan path folder tempat gambar-gambarnya berada
folder_path = 'D:/Datasets/analisis_citra/bali'

# Pastikan file hanya gambar (misalnya .jpg, .png)
valid_extensions = ['.jpg', '.jpeg', '.png']

# Ambil semua file dalam folder yang memiliki ekstensi valid
files = [f for f in os.listdir(folder_path) if os.path.splitext(f)[1].lower() in valid_extensions]

# Urutkan file (opsional, agar konsisten)
files.sort()

# Ganti nama file satu per satu
for idx, filename in enumerate(files, start=1):
    ext = os.path.splitext(filename)[1]  # Ambil ekstensi file (.jpg, .png, dll)
    new_name = f"bali_{idx}{ext}"
    old_path = os.path.join(folder_path, filename)
    new_path = os.path.join(folder_path, new_name)

    os.rename(old_path, new_path)
    print(f"Renamed: {filename} -> {new_name}")
```
